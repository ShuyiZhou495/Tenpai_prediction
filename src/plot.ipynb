```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("output.txt", "r") as f:
    lines = f.readlines()

get = lambda line: [float(line.split()[12]), float(line.split()[21])]
accs = np.array([get(line) for line in lines[1::2]])
epochs = np.arange(1, len(accs) + 1)

plt.figure()
plt.plot(epochs, accs[:, 0], 'r', label='Train')
plt.plot(epochs, accs[:, 1], 'b', label='Validation')
plt.legend()
plt.ylabel("Accuray")
plt.xlabel("#Epochs")
plt.show()
```
