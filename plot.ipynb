```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("ouput2.txt", "r") as f:
    lines = f.readlines()

get_acc = lambda line: [float(line.split()[6]), float(line.split()[15])]
get_loss = lambda line: [float(line.split()[9]), float(line.split()[18])]
accs = np.array([get_acc(line) for line in lines[1::2]])
losses = np.array([get_loss(line) for line in lines[1::2]])
epochs = np.arange(1, len(accs) + 1)

fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].plot(epochs, accs[:, 0], 'r', label='Train')
axes[0].plot(epochs, accs[:, 1], 'b', label='Validation')
axes[0].legend()
axes[0].set_ylabel("Categorical Accuracy")
axes[0].set_xlabel("#Epochs")

axes[1].plot(epochs, losses[:, 0], 'r', label='Train')
axes[1].plot(epochs, losses[:, 1], 'b', label='Validation')
axes[1].legend()
axes[1].set_ylabel("Categorical Cross Entropy Loss")
axes[1].set_xlabel("#Epochs")
plt.show()
```
